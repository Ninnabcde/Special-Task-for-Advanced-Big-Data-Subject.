```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
#Create a dataframe with Pyspark by calling the dataset csv file
diabetes = spark.read.csv("/content/drive/MyDrive/BigData Lanjut/Responsi/diabetes.csv", header=True, inferSchema=True)
diabetes.show()
```

```python
#Create a new dataframe by calling 2 columns namely 'Glucose' and 'Age' then save it in data!
newDataset = diabetes.select('Glucose', 'Age')
newDataset.show()
```

```python
#Display data but convert it first
newDataset = newDataset.withColumn('Glucose', newDataset.Glucose.cast('float'))
newDataset = newDataset.withColumn('Age', newDataset.Age.cast('float'))
```

```python
newDataset.describe('Glucose', 'Age').show()
```

```python
#Convert Dataframe to RDD with age as "key" and Glucose as "value"!
rdd=newDataset.rdd.map(tuple)
rdd=rdd.map(lambda x: [x[1],x[0]])
rdd.collect()
```

```python
#Show what ages are in the rdd data!
rdd.keys().collect()
```

```python
#How many cases are there if calculated from each age?
rdd.countByKey()
```
